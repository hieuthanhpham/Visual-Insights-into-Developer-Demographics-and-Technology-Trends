# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
```

```python
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query the ConvertedComp column data from the master table
QUERY = """
SELECT ConvertedComp
FROM master
"""
converted_comp_data = pd.read_sql_query(QUERY, conn)

# Step 3: Close the connection to free up resources (after the query has executed)
conn.close()

# Step 4: Drop any null values for more accurate plotting
converted_comp_data = converted_comp_data.dropna()

# Step 5: Plot a histogram
plt.figure(figsize=(10, 6))
plt.hist(converted_comp_data['ConvertedComp'], bins=30, edgecolor='k')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query the Age column data from the master table
QUERY = """
SELECT Age
FROM master
"""
age_data = pd.read_sql_query(QUERY, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Drop any null values to avoid issues in plotting
age_data = age_data.dropna()

# Step 5: Plot a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(age_data['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query the Age and WorkWeekHrs columns from the master table
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
age_workweek_data = pd.read_sql_query(QUERY, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Drop any rows with null values in either Age or WorkWeekHrs
age_workweek_data = age_workweek_data.dropna()

# Step 5: Plot a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(age_workweek_data['Age'], age_workweek_data['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()


```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python

# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query the WorkWeekHrs, CodeRevHrs, and Age columns from the master table
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
workweek_coderev_age_data = pd.read_sql_query(QUERY, conn)

# Step 3: Close the connection
conn.close()

# Step 4: Drop any rows with null values in WorkWeekHrs, CodeRevHrs, or Age
workweek_coderev_age_data = workweek_coderev_age_data.dropna()

# Step 5: Plot a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(workweek_coderev_age_data['WorkWeekHrs'], workweek_coderev_age_data['CodeRevHrs'], 
            s=workweek_coderev_age_data['Age'], alpha=0.5)

plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size by Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 1: List all tables in the database to identify the correct table for querying
tables_query = "SELECT name as Table_Name FROM sqlite_master WHERE type = 'table'"
tables = pd.read_sql_query(tables_query, conn)
tables

# Query the top 5 databases that respondents wish to learn next year
query = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""

# Execute the query to get the data
top_databases = pd.read_sql_query(query, conn)

# Close the connection as it's no longer needed
conn.close()

# Plot the pie chart if data was successfully retrieved
if not top_databases.empty:
    plt.figure(figsize=(8, 8))
    plt.pie(top_databases['Count'], labels=top_databases['DatabaseDesireNextYear'], 
            autopct='%1.1f%%', startangle=140)

    plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
    plt.show()
else:
    print("No data found in the specified column.")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import matplotlib.pyplot as plt

# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 2: Query median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
age_group_data = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Step 3: Calculate median values for WorkWeekHrs and CodeRevHrs
median_workweekhrs = age_group_data['WorkWeekHrs'].median()
median_coderevhrs = age_group_data['CodeRevHrs'].median()

# Step 4: Prepare data for stacked bar chart
labels = ['Age Group 30-35']
workweek_data = [median_workweekhrs]
coderev_data = [median_coderevhrs]

# Step 5: Plot stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(labels, workweek_data, label='Median WorkWeekHrs')
ax.bar(labels, coderev_data, bottom=workweek_data, label='Median CodeRevHrs')

# Add labels and title
ax.set_ylabel('Hours')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
ax.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Step 1: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get ConvertedComp and Age for respondents aged between 45 and 60
query = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Execute the query to get the data
compensation_data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Calculate the median ConvertedComp for each age group in Python
median_comp_data = compensation_data.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the line chart if data is available
if not median_comp_data.empty:
    plt.figure(figsize=(10, 6))
    plt.plot(median_comp_data['Age'], median_comp_data['ConvertedComp'], marker='o')
    plt.title('Median Converted Compensation for Ages 45 to 60')
    plt.xlabel('Age')
    plt.ylabel('Median Converted Compensation')
    plt.grid(True)
    plt.show()
else:
    print("No data available for the specified age range.")

```
